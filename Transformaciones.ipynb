```python
# importamos las librerias a utilizar
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
from IPython.display import clear_output
```

**Transformación Set de Datos Amazon:**

```python
df_amazon = pd.read_csv("..//MLOpsReviews/amazon_prime_titles.csv", sep=',' , header=0)
df_amazon.info()
```

1. Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
# Función lambda para agregar un caracter al inicio de cada valor en la columna 'Col1'
add_char = lambda x: 'a' + x

# Aplicamos la función a la columna 'Col1' usando el método apply()
df_amazon['show_id'] = df_amazon['show_id'].apply(add_char)
```

```python
# Imprimimos el DataFrame resultante
df_amazon.head()
```

2. Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”

```python
df_amazon['rating'] = df_amazon['rating'].fillna("G")
```

```python
count_df = df_amazon['rating'].value_counts().reset_index()
count_df
```

3. De haber fechas, deberán tener el formato AAAA-mm-dd

```python
df_amazon['date_added'] = pd.to_datetime(df_amazon['date_added'])
df_amazon['date_added'] = df_amazon['date_added'].dt.strftime('%Y-%m-%d')
df_amazon.head()
```

4. Los campos de texto deberán estar en minúsculas, sin excepciones

```python
# Aplicamos la función str.lower() a todos los valores del DataFrame utilizando el método applymap()
df_amazon = df_amazon.applymap(lambda x: x.lower() if type(x) == str else x)
df_amazon.head()
```

5. El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

Nota: 'str.extract()' se usa para extraer la cantidad de tiempo y la unidad de tiempo en dos columnas diferentes ('duration_int', 'duration_type'), utilizando una expresión regular que coincide con uno o más dígitos (\d+) seguidos de uno o más caracteres alfabéticos (\w+).


```python
# Utilizamos str.extract() para separar la cantidad de tiempo y la unidad de tiempo en dos columnas diferentes
df_amazon[['duration_int', 'duration_type']] = df_amazon['duration'].str.extract('(\d+) (\w+)')
df_amazon['duration_int'] = pd.to_numeric(df_amazon['duration_int'])
df_amazon.head()
```

```python
df_amazon.to_csv('TAmazon.csv', index=False)
```
